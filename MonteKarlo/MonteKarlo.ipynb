# Монте-карло
Методы Монте-Карло — это набор методов в математике для изучения случайных процессов. Случайных — это когда что-то в них происходит непредсказуемым образом, например:

- подбрасываем монетку;
- кидаем кубик;
- опускаем жетоны в ячейки со столбиками;
- ловим элементарные частицы;
- считаем столкновения молекул;
- и что угодно ещё, что происходит полностью случайно и что нельзя предсказать заранее.

Смысл методов Монте-Карло в том, чтобы использовать данные случайных событий, чтобы на их основе получить более-менее точные результаты каких-то других вычислений. Они не будут идеально и математически точными, но их уже будет достаточно, чтобы с ними полноценно работать. Иногда это проще и быстрее, чем считать всё по точным формулам.

## Симулятор
В этой программе симулируется игра в **Кости** - бросаются два кубика. 

В том случае, если выпадают равные значения, то игрок выигрывает, и его ставка (*step*) увеличивается в 4 раза. Если выпадают разные значения, то ставка забирается в пользу казино.

*Своё название метод получил в честь Монте-Карло — района Монако, где находится много казино с рулеткой, самым доступным источником случайных чисел в начале 20-го века.*

```python
# загрузим необходимые библиотеки
import matplotlib.pyplot as plt
import random
import numpy as np

# функция, которая проверяет результаты броска
def brosok ():
    br1=random.randint(1,6)
    br2=random.randint(1,6)
    if br1==br2:
        res='Win'
    else:
        res='Loos'
    return (res)

#функция, которая делает 5000 бросков, выводит график

def graph(s, sv, mg):
    k=1
    while k<5:
        w=[]
        summa=s
        stavka=sv
        mnogitel=mg
        r = np.round(np.random.rand(),1)
        g = np.round(np.random.rand(),1)
        b = np.round(np.random.rand(),1)
        for i in range(5000):
            w=brosok()
            if w=='Win':
                #в случаи выигрыша, ставка увеличивается в 4 раза
                summa=summa+(mnogitel*stavka)
            else:
                #в случаи проигрыша - игрок теряет ставку
                summa=summa-stavka
            plt.plot(i,summa, 
                     marker='.',
                     markersize=1,
                     color=[r,g,b])
            #очень редкий случай - все деньги проиграны до того, как было сделано 1000 бросков
            if summa<=3:
                break
        k=k+1



summa=int(input("У вас с собой сумма "))
stavka=int(input('Ваша ставка '))
mnogitel=int(input('Стартовый множитель казино '))
while mnogitel<=0:
    print ('Множитель казино должен быть больше 0')
    mnogitel=int(input('Стартовый множитель казино'))
print('__________________________')
print('Итак, мы начинаем')
print ('Ваша сумма', summa)
print('Ваша ставка', stavka)
print ('Множитель казино', mnogitel)
print('Бросаем 15 раз по 1000 бросков')
fig=plt.figure()
for g in range(5):
    print(mnogitel)
    graph(summa, stavka, mnogitel)
    mnogitel=mnogitel+1
    plt.show()
```

## Итак
Как видно, при ставке до х4 во время выигрыша - казино в 90% случаях остается в плюсе.

При х5 - возможны варианты

Игрок начинает выигрывать постоянно при ставке х6.
